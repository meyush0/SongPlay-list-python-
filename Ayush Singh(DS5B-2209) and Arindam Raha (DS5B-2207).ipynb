```python
#made by Ayush Singh(DS5B-2209) and Arindam Raha (DS5B-2207)

class SongLibrary:
    
    
    def __init__(self):
        self.songs = {
            "song1": {"artist": "Arjit singh", "year": "2020", "genre": "pop"},
            "song2": {"artist": "artist2", "year": "2018", "genre": "rock"},
            "song3": {"artist": "artist3", "year": "2019", "genre": "pop"},
            "song4": {"artist": "artist4", "year": "2021", "genre": "rap"}
        }
    
    def search(self):
        song_name = input("enter song name u want to search: ")
        if song_name in self.songs:
            return self.songs[song_name]
        else:
            print("the song is not in the list!!")
    
    def update(self, artist=None, year=None, genre=None):
        song_name=input("enter your song name!!!")
        if song_name in self.songs:
            if artist is not None:
                self.songs[song_name]["artist"] = artist
            if year is not None:
                self.songs[song_name]["year"] = year
            if genre is not None:
                self.songs[song_name]["genre"] = genre
            return True
        else:
            return False
    
    def delete(self):
        song_name=input("Enter the song name which you want to delete:-")
        if song_name in self.songs:
            del self.songs[song_name]
            return self.display()
        else:
            print("song in not in the playlist!!")
    
    def display(self):
        for song_name, song_info in self.songs.items():
            print(f"Song Name: {song_name}")
            print(f"\tArtist: {song_info['artist']}")
            print(f"\tYear: {song_info['year']}")
            print(f"\tGenre: {song_info['genre']}\n")

# Example usage
library = SongLibrary()

"""# Search for a song
print(library.search("song1"))  # {"artist": "artist1", "year": "2020", "genre": "pop"}
print(library.search("song5"))  # None

# Update a song
print(library.update("song2", artist="new artist"))  # True
print(library.search("song2"))  # {"artist": "new artist", "year": "2018", "genre": "rock"}

# Delete a song
print(library.delete("song3"))  # True
print(library.search("song3"))  # None
"""
"""# Display all songs
library.display()"""

#library.search()
library.delete()
library.search()
library.update()
library.display()
```
